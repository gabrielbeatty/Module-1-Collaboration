#This app will render a list of students who made the Deans List or the Honor Roll. 

```python
while True:
    lname = input("\nPlease enter student's last name or ZZZ to quit.")

    if lname.upper() == 'ZZZ':
        break  
    fname = input("Please enter student's first name: ")

    try:
        gpa = float(input("Please enter student's GPA (example: 3.5): "))
    except ValueError:
        print("Invalid GPA. Please enter a numerical value that includes a decimal.")
        continue

    if gpa >= 3.5:
        print(f"{fname} {lname} has made the Dean's List.")

    if gpa >= 3.25:
        print(f"{fname} {lname} has made the Honor Roll.")

print("End of Program")



```
